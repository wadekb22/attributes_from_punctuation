```python
import pandas as pd 
```

```python
df = pd.read_csv("stats_nourl.csv") 
# Preview the first 5 lines of the loaded data 
df.head()
```

```python
bins = 10
#labels = range(bins)
df['age_cat'] = pd.cut(df['age'],bins)
```

```python
df = df.drop(['age'],axis=1)
```

```python
# Merging Prob Between Punc (geometrically)
bin_size = 1
start = 0
num_cols = 40
eval_str = ""
while start+bin_size < num_cols:
    eval_str = "df['Prob " + str(start) + "-" + str(start+bin_size-1) + " Words Between Punc'] = "
    for j in range(bin_size):
        eval_str += "df['Prob "+ str(start+j) + " Words Between Punc'] + "
    eval_str = eval_str[0:-3]
    exec(eval_str)
    start = start+bin_size
    bin_size *= 2
#add remaining vals to final category
eval_str = "df['Prob " + str(start) + "-" + str(num_cols-1) + " Words Between Punc'] = "
for j in range(start,num_cols):
    eval_str += "df['Prob "+ str(j) + " Words Between Punc'] + "
eval_str = eval_str[0:-3]
exec(eval_str)
start = start+bin_size
bin_size *= 2
```

```python
# Removing old values of Prob Between Punc
num_cols = 40
eval_str = "df = df.drop(["
for i in range(num_cols):
    eval_str += "'Prob "+ str(i) + " Words Between Punc',"
eval_str = eval_str[0:-1]
eval_str += "],axis=1)"
exec(eval_str)
```
